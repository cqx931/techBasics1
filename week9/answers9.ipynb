```python
import pygame

# constants
SCREEN_WIDTH = 1200
SCREEN_HEIGHT = 300
BACKGROUND_COLOR = (255,255,255) # RGB

# key pressed states
left_pressed = False
right_pressed = False

class Dino:
    def __init__(self, pos_x:int, pos_y:int):
        img = pygame.image.load("../week8/dino.png")
        self.img = pygame.transform.scale(img, (100,100))
        # init position
        self.pos_x = pos_x
        self.pos_y = pos_y

    def tint(self):
        self.img.fill((0, 0, 200, 100), special_flags=pygame.BLEND_ADD)
        pass

    def animate(self, dr:str):
        direction = 1
        if dr == "left":
            direction = -1

        if self.pos_x < SCREEN_WIDTH:
           self.pos_x += direction * 5
        else:
           self.pos_x = 0

    def draw(self):
        screen.blit(self.img, (self.pos_x, self.pos_y))


# activate the pygame library
pygame.init()

# create the display surface object
# of specific dimension.
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# set the pygame window name
pygame.display.set_caption('image')

# Create one dino object at a start location
dino = Dino(0,100)

# Init the clock
clock = pygame.time.Clock()

flag = True

while flag:
    # ticking the clock
    clock.tick(60)

    if left_pressed:
        dino.animate("left")
    if right_pressed:
        dino.animate("right")


    # paint the screen with background color
    screen.fill(BACKGROUND_COLOR)
    # Using blit to copy image to screen at a specific location
    dino.draw()
    # refresh the display
    pygame.display.flip()

    # code you need to end pygame
    for event in pygame.event.get():

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                left_pressed = True
            if event.key == pygame.K_RIGHT:
                right_pressed = True

        if event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT:
                left_pressed = False
            if event.key == pygame.K_RIGHT:
                right_pressed = False

        if event.type == pygame.QUIT:
            flag = False

pygame.quit()
exit(0)
```

```python
import pygame

# constants
SCREEN_WIDTH = 1200
SCREEN_HEIGHT = 300
BACKGROUND_COLOR = (255,255,255) # RGB

key_states = {
    pygame.K_LEFT: False,
    pygame.K_RIGHT: False,
    pygame.K_a: False,
    pygame.K_d: False
}

class Dino:
    def __init__(self, pos_x:int, pos_y:int):
        img = pygame.image.load("../week8/dino.png")
        self.img = pygame.transform.scale(img, (100,100))
        self.original_img = self.img.copy()
        # init position
        self.pos_x = pos_x
        self.pos_y = pos_y

    def tint(self):
        self.img.fill((200, 0, 0, 100), special_flags=pygame.BLEND_ADD)

    def untint(self):
        self.img = self.original_img.copy()

    def animate(self, dr:str):
        direction = 1
        if dr == "left":
            direction = -1

        if self.pos_x < SCREEN_WIDTH:
           self.pos_x += direction * 5
        else:
           self.pos_x = 0

    def draw(self):
        screen.blit(self.img, (self.pos_x, self.pos_y))

    def is_closer_to_goal(self, another:"Dino"):
        return self.pos_x > another.pos_x

def handle_input():
    # Update key states based on Pygame events.
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            return False

        if event.type in (pygame.KEYDOWN, pygame.KEYUP):
            key_pressed = event.type == pygame.KEYDOWN
            if event.key in key_states:
                key_states[event.key] = key_pressed

    return True

def update(dino1, dino2):
    # Handle animation and tinting logic.
    if key_states[pygame.K_LEFT]:
        dino1.animate("left")
    if key_states[pygame.K_RIGHT]:
        dino1.animate("right")
    if key_states[pygame.K_a]:
        dino2.animate("left")
    if key_states[pygame.K_d]:
        dino2.animate("right")

    # Tinting based on who is ahead
    if dino1.is_closer_to_goal(dino2):
        dino1.tint()
        dino2.untint()
    elif dino2.is_closer_to_goal(dino1):
        dino2.tint()
        dino1.untint()
    else:
        dino1.untint()
        dino2.untint()
# activate the pygame library
pygame.init()

# create the display surface object
# of specific dimension.
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# set the pygame window name
pygame.display.set_caption('image')

# Create one dino object at a start location
dino = Dino(0,100)
dino2 = Dino(0,50)

# Init the clock
clock = pygame.time.Clock()

flag = True

while flag:
    # ticking the clock
    clock.tick(60)
    # update both dinos
    update(dino, dino2)
    # paint the screen with background color
    screen.fill(BACKGROUND_COLOR)
    # draw
    dino.draw()
    dino2.draw()
    # refresh the display
    pygame.display.flip()
    # event handling
    flag = handle_input()

pygame.quit()
exit(0)
```

```python
import pygame

# constants
SCREEN_WIDTH = 1200
SCREEN_HEIGHT = 300
BACKGROUND_COLOR = (255,255,255) # RGB
left_pressed = False
right_pressed = False


class Dino:
    def __init__(self, pos_x:int, pos_y:int):
        img = pygame.image.load("../week8/dino.png")
        self.img = pygame.transform.scale(img, (100,100))
        # init position
        self.pos_x = pos_x
        self.pos_y = pos_y

    def tint(self):
        self.img.fill((0, 0, 200, 100), special_flags=pygame.BLEND_ADD)
        pass

    def animate(self, dr:str):
        direction = 1
        if dr == "left":
            direction = -1

        if self.pos_x < SCREEN_WIDTH:
           self.pos_x += direction * 5
        else:
           self.pos_x = 0

    def draw(self):
        screen.blit(self.img, (self.pos_x, self.pos_y))

class DinoKid(Dino):

    def __init__(self, pos_x: int, pos_y: int):
        super().__init__(pos_x, pos_y)
        self.img = pygame.transform.scale(self.img, (50, 50))

    def follow(self, parent:Dino):
        self.pos_x = parent.pos_x - 40

# activate the pygame library
pygame.init()

# create the display surface object
# of specific dimension.
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# set the pygame window name
pygame.display.set_caption('image')

# Create one dino object at a start location
dino = Dino(50,100)
dinoKid = DinoKid(10,150)

# Init the clock
clock = pygame.time.Clock()

flag = True

while flag:
    # ticking the clock
    clock.tick(60)

    if left_pressed:
        dino.animate("left")
    if right_pressed:
        dino.animate("right")

    dinoKid.follow(dino)

    # paint the screen with background color
    screen.fill(BACKGROUND_COLOR)
    # Using blit to copy image to screen at a specific location
    dino.draw()
    dinoKid.draw()
    # refresh the display
    pygame.display.flip()

    # code you need to end pygame
    for event in pygame.event.get():

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                left_pressed = True
            if event.key == pygame.K_RIGHT:
                right_pressed = True

        if event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT:
                left_pressed = False
            if event.key == pygame.K_RIGHT:
                right_pressed = False

        if event.type == pygame.QUIT:
            flag = False

pygame.quit()
exit(0)
```
