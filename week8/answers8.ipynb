```python
class MensaCard:

    def __init__(self, name:str, balance:float):
        self.name = name
        self.balance = float(balance)

    def __str__(self):
        return f"This is a mensa card of {self.name}. It has a balance of {self.balance}"

    def deposit_money(self, amount:float):
        if amount < 0:
            raise ValueError("Please deposit a positive amount")
        else:
            self.balance += amount

    def eat_lunch(self, option:str):
        # Deal with option
        to_be_deducted = 0
        if option == "A":
            to_be_deducted = 2.3
            print("Ate lunch A")
        elif option == "B":
            to_be_deducted = 3.8
            print("Ate lunch B")
        else:
            print("Please provide only Option A or B")

        # Deal with balance
        if to_be_deducted <= self.balance:
            self.balance -= to_be_deducted
            return True
        else:
            print("Balance too low. Please top up.")
            return False

card = MensaCard("David", 50)
card.name = "David"
# card.balance = 50
print(card)
card.deposit_money(50)
print(card)
card.eat_lunch("B")
print(card)
```
